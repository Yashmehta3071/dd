```python
# Load data
data_1_env = pd.read_csv("data_1_env.csv")
data_2_env = pd.read_csv("data_2_env.csv")
data_3_crop = pd.read_csv("data_3_crop.csv")
```

* get unique values from each column to see data entry errors

```python
def get_uniques(df):
    u = []
    for col in df:
        u.append(f"{col} ({str(type(df[col][0]))}): {df[col].unique()}")
    display(u)
print("data_1_env")
get_uniques(data_1_env)
print("data_2_env")
get_uniques(data_2_env)
print("data_3_crop")
get_uniques(data_3_crop)
```

```python
# fix data_1_env 

# convert Avg Min Temp from Ferenheit to Celsius
data_1_env.insert(2, "Avg Min Temp *C", (data_1_env['Avg Min Temp *F'] - 32) / 1.8)
data_1_env = data_1_env.drop("Avg Min Temp *F", axis=1)

# convert Yr Rain inches to int to remove string issues 
data_1_env.insert(3, "Yr Rain mm", data_1_env['Yr Rain inches'] * 25.4)
data_1_env = data_1_env.drop("Yr Rain inches", axis=1)

```

```python
# universel replacemnet dict
replacement = {'No': 'N','no': 'N', '0': 'N', 0.: 'N', 'n': 'N', 'Yes': 'Y','yes': 'Y', '1': 'Y', 1.: 'Y', 'y' : 'Y'}
# Column to perform replacement on 
replacement_columns = ['Heat Wave', 'Dry Spell', 'Cold Wave', 'Wet Spell']
# Perform replacement on data_1_env
for column in replacement_columns:
    data_1_env[column] = data_1_env[column].replace(replacement)
# Perform replacement on data_2_env
for column in replacement_columns:
    data_2_env[column] = data_2_env[column].replace(replacement)
```

```python
# first deal with nan values
print("Table 1: Rows with NaN values")
display(data_3_crop[data_3_crop.isnull().any(axis=1)])
print("Table 2: Duplicate rows.")
display(data_3_crop[data_3_crop.duplicated(subset=['Year', 'Location'], keep=False)])
print("Table 3: Rows where Michael Jimenez is observer.")
display(data_3_crop[data_3_crop['Observer'] == "Michael Jimenez"])
print("Table 4: Rows where location is South Janethaven")
display(data_3_crop[(data_3_crop['Location'] == "South Janethaven")&((data_3_crop['Year']>=2008)|(data_3_crop['Year'].isnull()))])
```

* based on above Table 1 adn Table 2 we can see that there is duplicate rows for each row with NaN values. This means we can remove rows 7, 23, and 107
* based on above Table 3 we can see that Observer Michael Jimenez is only assigned to Location South Janethaven. we can only see, in Table 3 that if Loaction for row 112 replaced with South Janethaven, It is duplicate of row 111. We can remove row 112
* based on above Table 4 we can see that there's trend of having only 1 observation per year. For year 2010 and 2011 we have observations yet there is row 95 with NaN year. based on values from other columns, we can infer that row 95 is duplicate of row 94. We can remove row 95
* Above proves that there exist duplicate rows for all the NaN values thuse we can remove NaN rows.

```python
# remove rows with NaN values from data_3_crops
data_3_crop = data_3_crop.dropna()
```

```python
# Year from float to int 
data_3_crop['Year'] = data_3_crop['Year'].astype(int)
# universel replacemnet dict
replacement = {'no': 'N', 'No': 'N', 0.: 'N', 'Yes': 'Y', 1.: 'Y'}
# Column to perform replacement on 
replacement_columns = ['Irrigation', 'Crop Damage']
# Perform replacement 
for column in replacement_columns:
    data_3_crop[column] = data_3_crop[column].replace(replacement)
```
